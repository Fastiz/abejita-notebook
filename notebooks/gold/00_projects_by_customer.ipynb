```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("projects")
    .getOrCreate()
)
```

```python
from pyspark.sql import functions as f

projects = spark.read.parquet("/app/data/silver/projects")
customers = spark.read.parquet("/app/data/silver/customers")

project_stats = projects.groupBy("customer_id").agg(
    f.count("*").alias("project_count"),
)

join = project_stats.join(customers, on=project_stats["customer_id"] == customers["id"], how="inner")

join.select(
    customers["name"],
    join["project_count"],
).show()
```
