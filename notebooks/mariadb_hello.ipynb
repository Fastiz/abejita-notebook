```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("mariadb-hello")
    .config("spark.jars.packages", "org.mariadb.jdbc:mariadb-java-client:3.4.1")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("ERROR")
```

```python
JDBC_URL = "jdbc:mariadb://mariadb:3306/mydb"
JDBC_OPTS = {
    "user": "spark",
    "password": "sparkpassword",
    "driver": "org.mariadb.jdbc.Driver",
}

df = (
    spark.read.format("jdbc")
    .option("url", JDBC_URL)
    .option("query", "SELECT 'hello from mariadb' AS message")
    .options(**JDBC_OPTS)
    .load()
)
df.show()
```
