```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("projects")
    .getOrCreate()
)
```

```python
from pyspark.sql import functions as f

projects_bronze = spark.read.parquet("/app/data/bronze/tabProyecto")

projects = projects_bronze.select(
    f.expr("uuid()").alias("id"),
    f.col("name").alias("external_id"),
    f.col("creation").alias("created_at"),
    f.col("modified").alias("updated_at"),
    f.col("margen_total").alias("margin")
)
```

```python
projects.write.mode("overwrite").parquet("/app/data/silver/projects")
```
