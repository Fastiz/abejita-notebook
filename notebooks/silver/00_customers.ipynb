```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("customers")
    .getOrCreate()
)
```

```python
from pyspark.sql import functions as f

customers_bronze = spark.read.parquet("/app/data/bronze/tabCliente")

customers = customers_bronze.select(
    f.expr("uuid()").alias("id"),
    f.col("name").alias("name"),
    f.col("creation").alias("created_at"),
    f.col("modified").alias("updated_at"),
)
```

```python
customers.write.mode("overwrite").parquet("/app/data/silver/customers")
```
